```python
import pandas as pd
import glob
import config

config.ruta
ruta_directorio = "C:/Users/User/Downloads/OneDrive_icfes/REPORTES ICFES WEB/*.xls"
```

```python
# Lista para almacenar cada DataFrame
lista_dfs = []

# Iterar sobre cada archivo en el directorio que coincide con el patrón
for archivo in glob.glob(ruta_directorio):
    # Leer las columnas A y B hasta la fila 100 de cada archivo
    df = pd.read_excel(archivo, sheet_name="Información general", usecols="A:B", nrows=20)
    df = df.dropna()
    df_transposed = df.T.reset_index(drop=True)
    df_transposed.columns = df_transposed.iloc[0]
    df_final = df_transposed.drop(df_transposed.index[0])

    # Renombrar las columnas según el diccionario
    nombres_columnas = {
        "Fecha de presentación del examen": "fecha_examen",
        "Fecha de corte del SIMAT": "corte_simat",
        "Fecha de actualización de datos ": "actualizacion_datos",
        "Entidad territorial": "entidad_territorial",
        "Tipo de entidad": "tipo_entidad",
        "Número de municipios adscritos": "municipios_adscritos",
        "Número de establecimientos": "establecimientos"
    }
    df_final.rename(columns=nombres_columnas, inplace=True)

    # Agregar el DataFrame a la lista
    lista_dfs.append(df_final)

# Concatenar todos los DataFrames en uno solo
df_concatenado_0 = pd.concat(lista_dfs, ignore_index=True)

# Ruta donde deseas guardar el archivo CSV
ruta_csv = f"{config.ruta}/mun_info_general.csv"

# Guardar el DataFrame como un archivo CSV
df_concatenado_0.to_csv(ruta_csv, index=False)
```
